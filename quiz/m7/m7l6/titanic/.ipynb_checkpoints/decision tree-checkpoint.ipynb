```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline  
from sklearn.model_selection import train_test_split

Y = data['Survived']

features = data.drop(['Survived'], axis=1)
X = features.drop(columns=['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'PassengerId'])

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
class ModifiedLabelEncoder(LabelEncoder):

    def fit_transform(self, y, *args, **kwargs):
        return super().fit_transform(y).reshape(-1, 1)

    def transform(self, y, *args, **kwargs):
        return super().transform(y).reshape(-1, 1)
    
class OneHotEncoderEnhanced(OneHotEncoder):
    
    def fit_transform(self, df, *args, **kwargs):        
        super().fit(df)
        return self.transform(df)
    
    def transform(self, df, *args, **kwargs):
        cols = df.columns
        categories = self.categories_
        new_columns = ['.'.join([column, class_]) for categ, column in zip(categories, cols) for class_ in categ]        
        return pd.DataFrame(data=super().transform(df), index=df.index, columns=new_columns)
    
```

```python
def preprocess_data(X_train):

    preprocess_numerical = Pipeline(steps=[
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')),
    ])

    preprocess_categorical = Pipeline(steps=[
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        #('onehotencoder', OneHotEncoderEnhanced()) #this would add columns, so would not be good for pipeline
    ])

    numerical_cols = ['Age', 'Fare']
    categorical_cols = ['Sex', 'Embarked']

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', preprocess_numerical, numerical_cols),
            ('cat', preprocess_categorical, categorical_cols)
        ])

    preprocessor.fit_transform(X_train.copy())

    onehot = OneHotEncoderEnhanced(sparse=False)

    X_train_new = pd.DataFrame(data=preprocessor.fit_transform(X_train.copy()), index=X_train.index, 
                               columns=numerical_cols+categorical_cols)
    X_train_new = X_train_new.join(onehot.fit_transform(X_train_new[categorical_cols])).drop(columns=categorical_cols)
    return X_train_new
#preprocessor.transformers_[1][1].named_steps['onehotencoder'].get_feature_names()
```

```python
from sklearn.tree import DecisionTreeClassifier

X_train_transformed = preprocess_data(X_train)

model = DecisionTreeClassifier(random_state=0)
model.fit(X_train_transformed, y_train)
```

```python
from sklearn.metrics import accuracy_score

X_test_transformed = preprocess_data(X_test)

# Making predictions
y_train_pred = model.predict(X_train_transformed)
y_test_pred = model.predict(X_test_transformed)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
```

```python
model = DecisionTreeClassifier(random_state=0, max_depth=6, min_samples_leaf=6, min_samples_split=10 )
model.fit(X_train_transformed, y_train)

print('The training accuracy is', accuracy_score(y_train, model.predict(X_train_transformed)))
print('The test accuracy is', accuracy_score(y_test, model.predict(X_test_transformed)))
```

```python
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier(n_estimators=50, max_depth=5, min_samples_leaf=5, oob_score=True)
randomforest.fit(X_train_transformed, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_test_predicted = randomforest.predict(X_test_transformed)
print('accuracy_score:{}'.format(accuracy_score(y_test_predicted, y_test)))
print('precision_score:{}'.format(precision_score(y_test_predicted, y_test)))
print('recall_score:{}'.format(recall_score(y_test_predicted, y_test)))
print('f1_score:{}'.format(f1_score(y_test_predicted, y_test)))
```
